#### Distribuição das variaveis

```python
from scipy.stats import normaltest
```

```python
ver_a_rel = ["G3","G2","G1"]

for i in ver_a_rel:
    
    stats,pval = normaltest(dados_p[i])
    
    if pval > 0.05:
        
        print(i,": Distribuição Normal")
    else:
        print(i,": Distribuição Não Normal")
        
       
```

### Vizualização Estudantes

```python
### visualização das variaveis 
plt.figure(figsize = (17,12))

plt.subplot(221)
sns.color_palette("hls", 8)
sns.histplot(dados_p['freetime'], kde = True, color = "green")
plt.title('Histograma do Tempo Livre', fontsize = 20)


plt.subplot(222)
sns.histplot(dados_p['goout'], kde = True, color = "magenta")
plt.title('Histograma de Saída', fontsize = 20)

plt.subplot(223)
sns.histplot(dados_p['Dalc'], kde = True, color = "green")
plt.title('Histograma do Beber dia de Semana', fontsize = 20)


plt.subplot(224)
sns.histplot(dados_p['Walc'], kde = True, color = "magenta")
plt.title('Histograma de beber final de semana', fontsize = 20)

plt.show()
```

Em média, os estudantes têm um tempo livre consideravel e aproveitam para sair com amigos. Nos dias de semana, estes estudantes 
consomen muito pouco alcool. Observa-se que nos finais de semana, a maioria consome pouco alcool também.

## Conclusão

A Escola Gabriel Pereira tem (65%) alunos. A maioria dos alunos está concentrada em áreas urbanas. Grande parte dos alunos pretende cursar o ensino superior, a maioria frequentou a creche, mora com os pais e tem acesso à internet. 43% dos alunos escolhem a escola pelo curso, 23% optam por estar perto de casa. 48% dos alunos têm atividades escolares, a maioria tem apoio educacional.

Mas as notas G1,G2,G3 cescem com o tempo.

O sexo que consumiu mais bebidas alcoólicas foi o Masculino.
Os alunos 
de 22 anos pontuaram abaixo da média (G1,G2,G3) para o alto consumo de bebidas alcoólicas durante a semana e fins de semana. 

O Álcool prejudica a atenção e a concentração. Deve-se ter o bom senso para não abusar da bebida.

Há uma forte correlação entre as notas na primeiro, segundo e terceiro período. Os níveis aumentam com o tempo. É normal que os alunos precisem se ajustar no início.
